```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Preprocessing and Overview

### Subtask:
Convert the 'last_updated' column to datetime objects for time-based analysis and display basic information about the DataFrame, including data types and missing values.


**Reasoning**:
To fulfill the subtask, I will first convert the 'last_updated' column to datetime objects, handling any conversion errors by coercing them to NaT.
```python
#Data Overview
df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')
df.info()
df.describe()
```

```python
skewness_temp = df['temperature_celsius'].skew()
skewness_pm25 = df['air_quality_PM2.5'].skew()
skewness_humidity = df['humidity'].skew()

print(f"Skewness of Temperature (Celsius): {skewness_temp:.2f}")
print(f"Skewness of Air Quality PM2.5: {skewness_pm25:.2f}")
print(f"Skewness of Humidity: {skewness_humidity:.2f}")
```

```python
#Country wise Aggregation
country_avg_data = df.groupby('country').agg({
    'temperature_celsius': 'mean',
    'air_quality_PM2.5': 'mean'
}).reset_index()

country_avg_data.head()
```

```python
#highest average temperature
highest_temp_countries = country_avg_data.sort_values(by='temperature_celsius', ascending=False).head(10)
lowest_temp_countries = country_avg_data.sort_values(by='temperature_celsius', ascending=True).head(10)

print("Top 10 Countries with Highest Average Temperature:")
print(highest_temp_countries)
print("\nTop 10 Countries with Lowest Average Temperature:")
print(lowest_temp_countries)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='temperature_celsius', data=highest_temp_countries, hue='country', palette='viridis', legend=False)
plt.title('Top 10 Countries with Highest Average Temperature')
plt.xlabel('Country')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#lowest average temperature
plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='temperature_celsius', data=lowest_temp_countries, hue='country', palette='coolwarm', legend=False)
plt.title('Top 10 Countries with Lowest Average Temperature')
plt.xlabel('Country')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#highest average air quality
highest_pm25_countries = country_avg_data.sort_values(by='air_quality_PM2.5', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='air_quality_PM2.5', data=highest_pm25_countries, hue='country', palette='Reds_d', legend=False)
plt.title('Top 10 Countries with Highest Average Air Quality PM2.5')
plt.xlabel('Country')
plt.ylabel('Average PM2.5 (µg/m³)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#analomy detection using zscore and iqr
from scipy.stats import zscore

df['temperature_zscore'] = zscore(df['temperature_celsius'])
df['zscore_anomaly'] = (df['temperature_zscore'].abs() > 3)

print("Number of Z-score anomalies:", df['zscore_anomaly'].sum())
df[['temperature_celsius', 'temperature_zscore', 'zscore_anomaly']].head()
```

```python
Q1 = df['temperature_celsius'].quantile(0.25)
Q3 = df['temperature_celsius'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df['iqr_anomaly'] = (df['temperature_celsius'] < lower_bound) | (df['temperature_celsius'] > upper_bound)

print("Number of IQR anomalies:", df['iqr_anomaly'].sum())
df[['temperature_celsius', 'iqr_anomaly']].head()
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(df.index, df['temperature_celsius'], color='gray', label='Normal Points', alpha=0.6, s=10)

# Highlight Z-score anomalies
zscore_anomalies = df[df['zscore_anomaly']]
plt.scatter(zscore_anomalies.index, zscore_anomalies['temperature_celsius'], color='red', label='Z-score Anomalies', s=20)

# Highlight IQR anomalies
iqr_anomalies = df[df['iqr_anomaly']]
plt.scatter(iqr_anomalies.index, iqr_anomalies['temperature_celsius'], color='blue', label='IQR Anomalies', s=20, alpha=0.7)

plt.title('Temperature Celsius with Z-score and IQR Anomalies')
plt.xlabel('Data Point Index')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#correlation analysis and heatmap
numerical_cols = [
    'temperature_celsius',
    'humidity',
    'wind_kph',
    'pressure_mb',
    'precip_mm',
    'cloud',
    'air_quality_PM2.5',
    'air_quality_PM10',
    'uv_index',
    'feels_like_celsius'
]

correlation_matrix = df[numerical_cols].corr()
print("Correlation Matrix (first 5 rows and columns):")
print(correlation_matrix.head())
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Weather Variables')
plt.show()
```

```python
#time series analysis using rolling
df['country'].value_counts()
```

```python
df_location = df[df['country'] == 'Bulgaria'].copy()
df_location.head()
```

```python
df_location = df_location.sort_values(by='last_updated')
df_location.head()
```

```python
df_location['rolling_mean_temp'] = df_location['temperature_celsius'].rolling(window=24).mean()
df_location.head()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(df_location['last_updated'], df_location['temperature_celsius'], label='Original Temperature', alpha=0.7)
plt.plot(df_location['last_updated'], df_location['rolling_mean_temp'], label='Rolling Mean Temperature (24-period)', color='red')

plt.title('Temperature Trends with Rolling Mean in Bulgaria')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
